### Exploratory Data Analysis

> 1. Gender

```python
ax = sns.countplot(x = 'Gender',data = df, hue = 'Gender', palette=['brown','green'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Gender_wise_Sale = df.groupby(['Gender'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
Gender_wise_Sale
```

```python
sns.barplot(x= 'Gender', y = 'Sales' ,data = Gender_wise_Sale, hue='Gender')
```

> #### From the above analysis we can see that most of the buyers are Females(F) and they have more money to spend.

> 2. ##### Age

```python
ax = sns.countplot(x='Age Group', hue= 'Gender', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_by_Age = df.groupby(['Age Group'], as_index=False)['Sales'].sum().sort_values(by = 'Sales',ascending=False)

sns.barplot(x='Age Group', y= 'Sales', data= Sales_by_Age)
```

> #### From the analysis we can see that most of the buyers are of age group 26-35 and among them mostly are Females.

> 3. ##### State

```python
# Total number of orders from top 5 state
Order_per_State = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending=False).head(5)

sns.barplot(data=Order_per_State, x= 'State', y= 'Orders')
```

```python
# total sales amount per state
Sales_by_State = df.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(5)
sns.barplot(data= Sales_by_State, x= 'State', y='Sales')
```

> #### From the above graphs we can see that mostly orders are from UttarPradesh, Maharashtra and Karnataka respectively and same goes for sales too.

> 4. ##### Marital Status

```python
ax = sns.countplot(data=df,x='Marital_Status')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_by_Marital_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.barplot(data=Sales_by_Marital_status, x='Marital_Status', y='Sales',hue='Gender')
```

> #### from the above analysis we can observe that most of the buyers are married(women) and they have high purchasing power

> 5. Occupation

```python
ax = sns.countplot(data=df, x= 'Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_by_Occupation = df.groupby(['Occupation'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.barplot(data=Sales_by_Occupation,x='Occupation',y='Sales')
```

> ##### From the above graph we can observe that most of the buyers are working in IT Sector, Healthcare and Aviation.

> 6. ##### Product Category

```python
ax = sns.countplot(data=df,x='Product_Category')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_by_PC = df.groupby(['Product_Category'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(5)

sns.barplot(data=Sales_by_PC,x='Product_Category', y= 'Sales')
```

> ##### From the above analysis we can observe that most of the sold product are from Food, Clothing and Electronics Category.

### Conclusion

- Married Women age group 26-35 from UP,Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from category Food, Clothing and Electronics
